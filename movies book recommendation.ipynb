```python
import pandas as pd
import os
import surprise
from surprise.model_selection import train_test_split
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise import KNNWithMeans
import numpy as np
```

```python
rating = pd.read_csv("rating.csv")
rating.head()
```

```python
reader = Reader(rating_scale=(1,10))
data = Dataset.load_from_df(rating[['User ID','Movie ID','Rating']], reader)
```

```python
[trainset,testset] = train_test_split(data, test_size=0.15, shuffle=True)

recom = KNNWithMeans(k=50, sim_options={'name':'cosine','user_based':True})

recom.fit(data.build_full_trainset())      # using full training set and not doing k

test_pred = recom.test(testset)
RMSE = accuracy.rmse(test_pred)
```

```python
'''predict() function is used to compute the rating prediction for given user and item.'''


pre = recom.predict(1,6)
print("The rating value for item 6 by user 1 is", pre[3])


pre = recom.predict(2,2)
print("The rating value for item 2 by user 2 is", pre[3])


pre = recom.predict(2,4)
print("The rating value for item 4 by user 2 is", pre[3])


pre = recom.predict(3,1)
print("The rating value for item 1 by user 3 is", pre[3])


pre = recom.predict(4,5)
print("The rating value for item 5 by user 4 is", pre[3])

pre = recom.predict(5,2)
print("The rating value for item 2 by user 5 is", pre[3])
```

```python
# Que --> keep  a threshold of 4.5 and then display recommendations
```

```python
for i in pre:
    if pre[3]>4.5:
        print("recommendate")
    else:
        print("not recommendate")
```

```python
book = pd.read_csv("book_recom.csv")
book.head()
```

```python
book = book[['userID','bookTitle','bookRating']]
book.head(2)
```

```python
book_rating = book.pivot(index='bookTitle',columns='userID',values='bookRating').fillna(0)
book_rating
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import sklearn
```

### csr_matrix(): Compressed Sparse Row Matrix

```python
book_matrix = csr_matrix(book_rating.values)
book_matrix
```

```python
knn_model = NearestNeighbors(metric= 'cosine', algorithm= 'brute')
knn_model.fit(book_matrix)
```

* Find the similar (with maximum similarity -means lowest distance) books as that of the first book '1st to Die:A novel'
* The function  kneighbours() is used to find k number of neighbours of a point.
* We need to compute 6 neighbours from book1 to all other books. Note that the shortest distance possible now is 0, which is
    book1 to book1 itself. Hence, we need to take 6 neighbours, instead of 5.

```python
query_index1 = 0     #1st book

distances, indices = knn_model.kneighbors(book_rating.iloc[query_index1, :].values.reshape(1,-1),n_neighbors=6)
```

```python
print("Recommendation for {0}\n".format(book_rating.index[query_index1]))
for i in indices.flatten()[1:]:
    print(book_rating.index[i])
```
